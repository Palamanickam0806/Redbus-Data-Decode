```python
import pandas as pd
import numpy as np
from pathlib import Path

def process_transaction_data(transaction_file='train/transactions.csv', 
                           train_file='train.csv'):

    
    print("Loading transaction data...")
    df = pd.read_csv(transaction_file)
    
    print(f"Original data shape: {df.shape}")
    print(f"Columns: {df.columns.tolist()}")
    
    # Filter for DBD 15-30 range only
    print("Filtering data for DBD 15-30 range...")
    df_filtered = df[(df['dbd'] >= 15) & (df['dbd'] <= 30)].copy()
    print(f"Filtered data shape: {df_filtered.shape}")
    
    print("Grouping by route and DOJ...")
    
    # Group by route (srcid, destid) and doj to create time series
    grouped = df_filtered.groupby(['doj', 'srcid', 'destid'])
    
    train_data = []
    
    for (doj, srcid, destid), group in grouped:
        # Sort by DBD in descending order (30 to 15)
        group_sorted = group.sort_values('dbd', ascending=False)
        
        # Get the final seat count (at DBD 15, which is minimum DBD)
        final_seat_count = group_sorted[group_sorted['dbd'] == group_sorted['dbd'].min()]['cumsum_seatcount'].iloc[0]
        
        # Create time series data - one row per DBD
        for _, row in group_sorted.iterrows():
            train_data.append({
                'doj': doj,
                'srcid': srcid,
                'destid': destid,
                'dbd': row['dbd'],
                'srcid_region': row['srcid_region'],
                'destid_region': row['destid_region'],
                'srcid_tier': row['srcid_tier'],
                'destid_tier': row['destid_tier'],
                'cumsum_seatcount': row['cumsum_seatcount'],
                'cumsum_searchcount': row['cumsum_searchcount'],
                'final_seatcount': final_seat_count
            })
    
    # Create DataFrame
    print("Creating training DataFrame...")
    train_df = pd.DataFrame(train_data)
    
    # Save file
    print(f"Saving train data to {train_file}...")
    train_df.to_csv(train_file, index=False)
    print(f"Train data shape: {train_df.shape}")
    
    # Display sample data
    print("\nSample train data:")
    print(train_df.head(10))
    
    # Data summary
    print(f"\nData Summary:")
    print(f"Unique routes: {train_df[['doj', 'srcid', 'destid']].drop_duplicates().shape[0]}")
    print(f"Date range: {train_df['doj'].min()} to {train_df['doj'].max()}")
    print(f"DBD range: {train_df['dbd'].min()} to {train_df['dbd'].max()}")
    print(f"Total records: {len(train_df)}")
    
    return train_df

def analyze_data_quality(train_df):
    """
    Analyze the quality and completeness of the processed data.
    """
    print("\n" + "="*50)
    print("DATA QUALITY ANALYSIS")
    print("="*50)
    
    # Check for missing values
    print("\nMissing values:")
    print(train_df.isnull().sum())
    
    # Check DBD distribution
    print(f"\nDBD distribution:")
    print(train_df['dbd'].value_counts().sort_index())
    
    # Check routes with incomplete time series
    route_counts = train_df.groupby(['doj', 'srcid', 'destid'])['dbd'].count()
    incomplete_routes = route_counts[route_counts < 16]  # Should have 16 records (DBD 15-30)
    
    print(f"\nRoutes with incomplete time series (< 16 records): {len(incomplete_routes)}")
    if len(incomplete_routes) > 0:
        print("Sample incomplete routes:")
        print(incomplete_routes.head())
    
    # Check for duplicate entries
    duplicates = train_df.duplicated(subset=['doj', 'srcid', 'destid', 'dbd']).sum()
    print(f"\nDuplicate route-dbd combinations: {duplicates}")
    
    return train_df


# Main execution
if __name__ == "__main__":
    try:       
        # Process the data
        train_df = process_transaction_data()
        
        # Analyze data quality
        analyze_data_quality(train_df)
        
        print("\n" + "="*50)
        print("PROCESSING COMPLETE!")
        print("="*50)
        print("File created:")
        print("- train.csv: Time series training data")
        print("\nYou can now use train.csv for time series forecasting model training.")
        
    except Exception as e:
        print(f"Error: {e}")
        import traceback
        traceback.print_exc()
```

```python
import pandas as pd

def update_final_seat_count(train_file='train.csv', 
                           metadata_file='train/train.csv',
                           output_file='train.csv'):
    """
    Update final_seatcount in train.csv using data from train_metadata.csv
    
    Parameters:
    - train_file: Training data CSV file
    - metadata_file: Metadata CSV with final seat counts
    - output_file: Updated training data CSV file
    """
    
    print("Loading train data...")
    train_df = pd.read_csv(train_file)
    print(f"Train data shape: {train_df.shape}")
    
    print("Loading metadata...")
    metadata_df = pd.read_csv(metadata_file)
    print(f"Metadata shape: {metadata_df.shape}")
    print(f"Metadata columns: {metadata_df.columns.tolist()}")
    
    # Display sample of metadata to understand structure
    print("\nSample metadata:")
    print(metadata_df.head())
    
    # Create merge key based on common columns
    # Assuming metadata has columns: doj, srcid, destid, final_seatcount
    merge_columns = ['doj', 'srcid', 'destid']
    
    # Check if merge columns exist in both dataframes
    missing_cols_train = [col for col in merge_columns if col not in train_df.columns]
    missing_cols_metadata = [col for col in merge_columns if col not in metadata_df.columns]
    
    if missing_cols_train:
        print(f"Warning: Missing columns in train data: {missing_cols_train}")
    if missing_cols_metadata:
        print(f"Warning: Missing columns in metadata: {missing_cols_metadata}")
    
    # Perform the merge
    print(f"\nMerging on columns: {merge_columns}")
    
    # First, drop the existing final_seatcount column from train data
    if 'final_seatcount' in train_df.columns:
        train_df = train_df.drop('final_seatcount', axis=1)
    
    # Merge with metadata to get updated final_seatcount
    train_updated = train_df.merge(
        metadata_df[merge_columns + ['final_seatcount']], 
        on=merge_columns, 
        how='left'
    )
    
    print(f"Updated train data shape: {train_updated.shape}")
    
    # Check for missing final_seatcount values after merge
    missing_count = train_updated['final_seatcount'].isnull().sum()
    print(f"Missing final_seatcount after merge: {missing_count}")
    
    if missing_count > 0:
        print("Routes with missing final_seatcount (sample):")
        missing_routes = train_updated[train_updated['final_seatcount'].isnull()][merge_columns].drop_duplicates().head()
        print(missing_routes)
    
    # Save updated file
    print(f"\nSaving updated data to {output_file}...")
    train_updated.to_csv(output_file, index=False)
    
    print("\nSample updated data:")
    print(train_updated[['doj', 'srcid', 'destid', 'dbd', 'cumsum_seatcount', 'final_seatcount']].head(10))
    
    # Summary statistics
    print(f"\nSummary:")
    print(f"Total records: {len(train_updated)}")
    print(f"Unique routes: {train_updated[merge_columns].drop_duplicates().shape[0]}")
    print(f"Final seat count range: {train_updated['final_seatcount'].min()} - {train_updated['final_seatcount'].max()}")
    
    return train_updated

def verify_update(original_file='train.csv', updated_file='train_updated.csv'):
    """
    Verify the update by comparing original and updated files
    """
    print("Verifying update...")
    
    original_df = pd.read_csv(original_file)
    updated_df = pd.read_csv(updated_file)
    
    print(f"Original final_seatcount unique values: {original_df['final_seatcount'].nunique()}")
    print(f"Updated final_seatcount unique values: {updated_df['final_seatcount'].nunique()}")
    
    print(f"Original final_seatcount range: {original_df['final_seatcount'].min()} - {original_df['final_seatcount'].max()}")
    print(f"Updated final_seatcount range: {updated_df['final_seatcount'].min()} - {updated_df['final_seatcount'].max()}")
    
    # Compare a sample route
    sample_route = updated_df[['doj', 'srcid', 'destid']].drop_duplicates().iloc[0]
    print(f"\nSample route comparison:")
    print(f"Route: {sample_route.values}")
    
    original_sample = original_df[
        (original_df['doj'] == sample_route['doj']) & 
        (original_df['srcid'] == sample_route['srcid']) & 
        (original_df['destid'] == sample_route['destid'])
    ]['final_seatcount'].iloc[0]
    
    updated_sample = updated_df[
        (updated_df['doj'] == sample_route['doj']) & 
        (updated_df['srcid'] == sample_route['srcid']) & 
        (updated_df['destid'] == sample_route['destid'])
    ]['final_seatcount'].iloc[0]
    
    print(f"Original final_seatcount: {original_sample}")
    print(f"Updated final_seatcount: {updated_sample}")
    print(f"Changed: {'Yes' if original_sample != updated_sample else 'No'}")

# Main execution
if __name__ == "__main__":
    try:
        # Update final seat count
        updated_df = update_final_seat_count()
        
        # Verify the update
        # verify_update()
        
        print("\n" + "="*50)
        print("UPDATE COMPLETE!")
        print("="*50)
        print("File created: train_updated.csv")
        print("Final seat counts have been updated from train_metadata.csv")
        
    except Exception as e:
        print(f"Error: {e}")
        import traceback
        traceback.print_exc()
```
